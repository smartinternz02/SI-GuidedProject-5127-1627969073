```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
```

```python
ct = ColumnTransformer([("one",OneHotEncoder(),[3])],remainder = "passthrough")
x = ct.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2,random_state = 0)
```

```python
from sklearn.preprocessing import StandardScaler
s = StandardScaler()
x_train = s.fit_transform(x_train)
x_test = s.fit_transform(x_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
regression = Sequential()
```

```python
# input
regression.add(Dense(units=6, kernel_initializer="random_uniform", activation="relu"))
#hidden layers
regression.add(Dense(units=12, kernel_initializer = "random_uniform", activation="relu"))
regression.add(Dense(units=12, kernel_initializer = "random_uniform", activation="relu"))
regression.add(Dense(units=12, kernel_initializer = "random_uniform", activation="relu"))
# output layers
regression.add(Dense(units=1, kernel_initializer = "random_uniform", activation="relu"))
```

```python
regression.compile(optimizer="rmsprop", loss="mse" ,metrics=["mse"])
```

```python
regression.fit(x_train, y_train, epochs = 1000)
```

```python
ypred = regression.predict(x_test)
```
